```python
from tkinter import *
from googletrans import Translator, LANGUAGES
from threading import Timer
translator = Translator()
on_off = False


def dark_on_off():
    global on_off
    if not on_off :
        lCanvas.config(bg="#00007f")
        rCanvas.config(bg="#D3D5D4")
        lCanvas.itemconfig(title, fill="#d1d9e1");lCanvas.itemconfig(subtitle1, fill="#d1d9e1")
        rCanvas.itemconfig(title, fill="#00007f");rCanvas.itemconfig(subtitle, fill="#00007f");rCanvas.itemconfig(subtitle1, fill="#00007f")
        textentry.config(bg="#dbdbdb");textentry1.config(bg="#dbdbdb")
        w.config(fg='#fff');w.config(bg='#00007f')
        dark_mode_bttn.config(image=DM_on_img, bg="#D3D5D4")
        
        on_off = True
    else :
        lCanvas.config(bg="#267dff")
        rCanvas.config(bg="#fff")
        lCanvas.itemconfig(title, fill="white");lCanvas.itemconfig(subtitle1, fill="white")
        rCanvas.itemconfig(title, fill="#267dff");rCanvas.itemconfig(subtitle, fill="#267dff");rCanvas.itemconfig(subtitle1, fill="#267dff")
        textentry.config(bg="white");textentry1.config(bg="white")
        w.config(fg='#000');w.config(bg='#fff')
        dark_mode_bttn.config(image=DM_off_img, bg="white")        
        on_off = False

window = Tk()
window.geometry("1000x700")
window.resizable(0,0)
window.iconbitmap('translation.ico')

# Listen to Entry input
msgToTrad = StringVar()
msgToTrad.trace("w", lambda name, index,  mode, msgToTrad = msgToTrad: callback())

# selected Language
selLang = StringVar()
selLang.set("arabic")
selLang.trace("w", lambda name, index, mode, selLang = selLang: callback())

# TraslantedMsg
translatedMsg = StringVar()

# Toutes les langues et leur code ISO
invLANGUAGES = dict((v, k) for k,v in LANGUAGES.items())
def chrono_update(args):
    chrono = Timer(3, getInput)
    chrono.start()
    
def getInput():
    msgToTrad.set(textentry1.get("1.0","end-1c"))

def callback() :
    msg = msgToTrad.get()
    if len(msg)>=1:
        selectedLang = selLang.get()
        translated = translator.translate(msg, dest=invLANGUAGES[selectedLang])
        textentry.delete("1.0", "end")
        textentry.insert("1.0", translated.text)
        rCanvas.itemconfigure(subtitle1, text="Langue Detectee : {}".format(LANGUAGES[translated.src]))

window.rowconfigure(0, weight=1)
window.columnconfigure(0, weight=1)
#image

DM_off_img = PhotoImage(file = "DM off.png")
DM_on_img = PhotoImage(file = "DM on.png")

# Colonne gauche
lCanvas = Canvas(window, bg="#267dff", width="500", height="700", highlightthickness=0)
lCanvas.place(x=0, y=0)

title = lCanvas.create_text(30, 100, anchor = "nw", font="Lato 21 bold", fill="white", width="400", text="Bienvenue dans notre traducteur Tkinter")

subtitle = lCanvas.create_text(30, 200, anchor ="nw", font="Lato 18", fill="white", width="400", text="Veuillez insérer n'importe quel message, et la langue de ce dernier serait automatiquement détectée puis votre texte sera traduit à votre langue souhaitée.")

# Colonne droite
rCanvas = Canvas(window, bg="#fff", width="500", height="700", highlightthickness=0)
rCanvas.place(x=500, y=0)

title = rCanvas.create_text(30, 50, anchor = "nw", font="Lato 24 bold", fill="#267dff",
width="400", text="Veuillez insérer votre message:")

subtitle1 = rCanvas.create_text(30, 150, anchor = "nw", font="Lato 16", fill="#267dff", width="400")
rCanvas.itemconfigure(subtitle1, text="Langue Detectee :")

textentry1 = Text(rCanvas, font="Lato 16", selectborderwidth="0px", borderwidth=0, highlightcolor="#267dff", highlightbackground="#d8e5f9", bg="#fbfbfb")
textentry1.bind('<KeyRelease>', lambda *args: chrono_update(args))
rCanvas.create_window(30, 180, window=textentry1, height=150, width=440, anchor="nw")


w = OptionMenu(rCanvas, selLang, *LANGUAGES.values())
w.config(bg='#fff',cursor="hand2")
rCanvas.create_window(30, 350, window=w, height=40, width=150, anchor="nw")

subtitle = rCanvas.create_text(30, 450, anchor = "nw", font="Lato 16", fill="#267dff", width="400", text="Message Traduit :")

textentry = Text(rCanvas, font="Lato 16", selectborderwidth="0px", borderwidth=0, highlightcolor="#267dff", highlightbackground="#d8e5f9", bg="#fbfbfb")
rCanvas.create_window(30, 480, window=textentry, height=150, width=440, anchor="nw")

dark_mode_bttn = Button(rCanvas, image=DM_off_img, bg="#fff", borderwidth=0, cursor="hand2", command=dark_on_off)
rCanvas.create_window(500, 5, window=dark_mode_bttn, height=40, width=40, anchor="ne")


window.mainloop()
```
